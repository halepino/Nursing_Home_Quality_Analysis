**Section 2:** My next goal will be to create a table of sum data at the state level. I will use sum data at the state level as this seems more appropriate for population characteristics as opposed to aggregate values.

```python
# Create a table of sum values by state
# subset columns that include values to be summed (not percentages)
# Start by getting names and indeces of all columns

for col in people.columns:
    index_no = people.columns.get_loc(col)
    print(index_no , col )
```

```python
# Transformation [4]
# Select columns to be summed and include state value to groupby
# State, All race columns, count for 65+ older citizens, and land area of each county.
state1 = people.iloc[:,[0,1,2,3,7,8,9,10,11,12,18]]

# Check dataframe
state1.head()
```

```python
# Transformation [5]
# Calculate counts for percentage columns by using population value.
# Percentage columns desired: Population, population Aged under 18, over 65, and all Education columns
# Subset these columns to transform and join
state2 = people.iloc[:,[0,1,3, 5, 6, 13, 14, 15, 16, 17]]

# Check dataframe
state2.head()
```

```python
state2 = pd.DataFrame(state2)
```

```python
# Transformation [6]
# Transform columns to counts
# Divide percentages by 100 and multiply by state total population to get counts
state2.loc[:,'Under18'] = round((state2.loc[:,'Under18'] / 100) * state2.loc[:,'Population'])
state2.loc[:,'Over65'] = round((state2.loc[:,'Over65'] / 100) * state2.loc[:,'Population'])
state2.loc[:,'UnderHS']= round((state2.loc[:,'UnderHS'] / 100) * state2.loc[:,'Population'])
state2.loc[:,'Ed_HSDiploma'] = round((state2.loc[:,'Ed_HSDiploma'] / 100) * state2.loc[:,'Population'])
state2.loc[:,'Ed_SomeCollege'] = round((state2.loc[:,'Ed_SomeCollege'] / 100) * state2.loc[:,'Population'])
state2.loc[:,'Ed_Assoc'] = round((state2.loc[:,'Ed_Assoc'] / 100) * state2.loc[:,'Population'])
state2.loc[:,'Ed_Degree'] = round((state2.loc[:,'Ed_Degree'] / 100) * state2.loc[:,'Population'])
```

```python
state2.head()
```

```python
#Transformation [7]
# Join the tables to create a complete table of columns with counts
# Check shapes of both tables
print(state1.shape, state2.shape)
```

```python
# Merge tables on FIPS code 
state = state1.merge(state2, how = 'inner', on = 'FIPS')

state.head()
```
